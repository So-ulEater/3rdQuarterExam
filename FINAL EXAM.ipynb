```python
import numpy as np

# Define the matrix A and vector b
A = np.array([[14, 8, 1],
              [1, 5, 18],
              [13, 18, 18]])

b = np.array([13, 6, 1])

# Solve using Matrix Inverse Method
try:
    A_inv = np.linalg.inv(A)         # Compute A^-1
    x_manual = np.dot(A_inv, b)      # x = A^-1 * b
    print("\nSolution using Matrix Inverse Method:")
    print(x_manual)
except np.linalg.LinAlgError:
    print("\nMatrix A is singular and cannot be inverted.")

# Solve using Gaussian Elimination
x_gaussian = np.linalg.solve(A, b)   # Direct Gaussian elimination
print("\nSolution using Gaussian Elimination:")
print(x_gaussian)

# Verification Step: Compute Ax and compare with b
b_check_inverse = np.dot(A, x_manual)    # Recompute b using A and x_manual
b_check_gaussian = np.dot(A, x_gaussian) # Recompute b using A and x_gaussian

print("\nVerification (Ax = b) for Inverse Method:")
print(b_check_inverse)

print("\nVerification (Ax = b) for Gaussian Elimination:")
print(b_check_gaussian)

# Final Check for Consistency
print("\nAre both solutions the same?")
print(np.allclose(x_manual, x_gaussian))
```
